```python
ls=[]
for i in range(1,51):
    base_url = 'https://books.toscrape.com/catalogue/page-{}.html'.format(i)
    request = requests.get(base_url)
    soup = bs4.BeautifulSoup(request.content,'lxml')
    data = soup.select('.product_pod')

    for book in data:
        if len(book.select('star-rating.Two'))!=0:
            book_title = book.select('a')[1]['title']
            ls.append(book_title)

ls
```
